# Data Reading

```python
df = spark.read.format('parquet')\
            .option('inferSchema',True)\
            .load('abfss://bronze@carprojectdatalake.dfs.core.windows.net/rawdata')  
```

```python
display(df)
```

#Data transformation

```python
from pyspark.sql.functions import *
```

```python
df=df.withColumn('model_category',split(col('Model_ID'),'-')[0])
```

```python
df=df.withColumn('Year',split(col('Date'),'-')[0])
```

```python
df.display()
```

```python
df=df.withColumn('RevPerUnit',col('Revenue')/col('Units_sold'))
df.display()
```

#Ad-Hoc

```python
df.groupBy('Year','Branch_Name').agg(sum('Units_sold').alias('Total_Units')).sort('Year','Total_Units',ascending=[1,0]).display()
```

#Data Writing

```python
df.write.format('parquet')\
    .mode('overwrite')\
    .option('path','abfss://silver@carprojectdatalake.dfs.core.windows.net/carsales')\
        .save()    
```
